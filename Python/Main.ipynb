```python
# Program 2.1

def cal():
    print("Simple Calculator")
    num1 = float(input("Enter First Number: "))
    num2 = float(input("Enter Second Number: "))
    
    print("Select an Operation")
    print("1. Add")
    print("2. Divide")
    
    choice = input("Enter 1/2")
    if choice == '1':
            print(f"Result: {num1} + {num2} = {num1 + num2}") 
    elif choice == '2':
        if num2 !=0:
            print(f"Result: {num1}/{num2}={num1 / num2}")
        else:
            print("Error!! Division by 0")
    else:
        print("invalid Input!!!")
cal()

# # END OF PROGRAM 2.1 # #
```

```python
# Program 2.2

def temp_conv():
    print("Temperature Converter")
    temp = float(input("Enter Temperature in Celsius"))
    print("Convert to: ")
    print("1. Celsius to Fahrenheit")
    print("2. Fahrenheit to Celsius")
    choice = input("Enter 1/2")
    if choice == '1':
        farh=(temp * 9/5) + 32
        print(f"{temp} Celsius is {farh }Fahrenheit")
    elif choice == '2':
        cel=(temp-32) * 5/9
        print(f"{temp} Fahrenheit is {cel} Celsius")
    else:
        print("Invalid input")
temp_conv()

# # END OF PROGRAM 2.2 # #
```

```python
# Program 2.3

def larg():       
    print("Find the Largest 3 Numbers")
    num1= float(input("Enter First Number: "))
    num2= float(input("Enter Second Number: "))
    num3= float(input("Enter Third Number: "))
    
    if num1>=num2 and num1>=num3:
        lar=num2
    elif num2>=num1 and num2>=num3:
        lar=num2
    else:
        lar=num3
        
    print(f"The largest number is: {lar}")
larg()

# # END OF PROGRAM 2.3 # #
```

```python

# # PROGRAM 3.1 # #

def fact(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * fact(n-1)

num = int(input("Enter The number: "))
print(f"The factorial of {num} is {fact(num)}")
# # END OF PROGRAM 3.1 # #
```

```python
# # PROGRAM 3.2 # #

def word_count(sent):
    word = sent.split()
    return len(word)

sent = input("Enter a sentence: ")
print(f"The number of words in the sentence is: {word_count(sent)}")

# # END OF PROGRAM 3.1 # #


# u = input("Enter A sentence: ")
# c = u.split()
# print(c)
```
